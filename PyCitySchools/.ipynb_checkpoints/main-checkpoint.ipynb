```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

# District Summary

    Calculate the total number of schools

    Calculate the total number of students

    Calculate the total budget

    Calculate the average math score

    Calculate the average reading score

    Calculate the percentage of students with a passing math score (70 or greater)

    Calculate the percentage of students with a passing reading score (70 or greater)

    Calculate the percentage of students who passed math and reading (% Overall Passing)

    Create a dataframe to hold the above results

    Optional: give the displayed data cleaner formatting

```python
# Print the total numbers of school to the screen
total_school = len(school_data_complete["school_name"].unique())

# Print the total numbers of students to the screen
total_students = len(school_data_complete["student_name"])

# Print the total budget of school to the screen
total_budget = (school_data["budget"]).sum()

# Print the average math score of school data complete to the screen
average_mathscore = round((school_data_complete["math_score"]).mean(),6)

# Print the average reading score of school data complete to the screen
average_readingscore = round((school_data_complete["reading_score"]).mean(),6)

# Print the passing percentage maths of school data complete to the screen
passing_percent_maths = round((len(school_data_complete.loc[school_data_complete["math_score"] >= 70,:])/total_students)*100,6)

# Print the passing percentage reading of school data complete to the screen
passing_percent_reading = round((len(school_data_complete.loc[school_data_complete["reading_score"] >= 70,:])/total_students)*100,6)

# Print the passing percentage overall of school data complete to the screen
passing_percent_overall = round((len(school_data_complete.loc[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70),:])/total_students)*100,6)
```

```python
# Creating a dataframe to hold the above results
school_df = pd.DataFrame({ "Total Schools" : [total_school],
                          "Total Students" : [total_students],
                          "Total Budget" : [total_budget],
                          "Average Math Score" : [average_mathscore],
                          "Average Reading Score" : [average_readingscore],
                          "% Passing Math" : [passing_percent_maths],
                          "% Passing Reading" : [passing_percent_reading],
                          "% Overall Passing" : [passing_percent_overall]
    
    
})
school_df
```

# School Summary

## Create an overview table that summarizes key metrics about each school, including:

    School Name
    School Type
    Total Students
    Total School Budget
    Per Student Budget
    Average Math Score
    Average Reading Score
    % Passing Math
    % Passing Reading
    % Overall Passing (The percentage of students that passed math and reading.)
    Create a dataframe to hold the above results

```python
reduce_school_df = school_data_complete[["school_name","type","student_name","reading_score","math_score","size","budget"]]
# reduce_school_df.head()
```

```python
grouped_school_df = reduce_school_df.groupby(["school_name","type"])
# grouped_school_df.head()
```

```python
ttl_students = grouped_school_df["student_name"].count()

ttl_budget = grouped_school_df["budget"].mean()

ttl_math_score = grouped_school_df["math_score"].sum()
ttl_reading_score = grouped_school_df["reading_score"].sum()

math_passing_df = reduce_school_df.loc[(reduce_school_df["math_score"] >= 70), :]
grouped_math_df = math_passing_df.groupby(["school_name","type"])
ttl_math_pass_cnt = grouped_math_df["student_name"].count()

reading_passing_df = reduce_school_df.loc[(reduce_school_df["reading_score"] >= 70), :]
grouped_reading_df = reading_passing_df.groupby(["school_name","type"])
ttl_reading_pass_cnt = grouped_reading_df["student_name"].count()

overall_passing_df = reduce_school_df.loc[(reduce_school_df["math_score"] >= 70) & (reduce_school_df["reading_score"] >= 70) , :]
grouped_overall_passing_df = overall_passing_df.groupby(["school_name","type"])
ttl_overall_pass_cnt = grouped_overall_passing_df["student_name"].count()


school_summ_df = pd.DataFrame({"Total Students" : ttl_students,
                                "Total Budget" : ttl_budget,
                              "Total Maths Score":ttl_math_score,
                              "Total Reading Score":ttl_reading_score,
                               "Total Math Pass Count":ttl_math_pass_cnt,
                               "Total Reading Pass Count":ttl_reading_pass_cnt,
                               "Total Overall Pass Count":ttl_overall_pass_cnt
                              })

school_summ_df["Per Student Budget"]  = school_summ_df["Total Budget"]/school_summ_df["Total Students"]
school_summ_df["Average Math Score"] = school_summ_df["Total Maths Score"]/school_summ_df["Total Students"]
school_summ_df["Average Reading Score"] = school_summ_df["Total Reading Score"]/school_summ_df["Total Students"]
school_summ_df["% Passing Math"] = (school_summ_df["Total Math Pass Count"]/school_summ_df["Total Students"])*100
school_summ_df["% Passing Reading"] = school_summ_df["Total Reading Pass Count"]/school_summ_df["Total Students"]*100
school_summ_df["% Overall Passing"] = school_summ_df["Total Overall Pass Count"]/school_summ_df["Total Students"]*100

school_summ_df = school_summ_df[["Total Students","Total Budget","Per Student Budget","Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]]
school_summ_df
```

# Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
    * Average Math Score
    * Average Reading Score
    * % Passing Math
    * % Passing Reading
    * Overall Passing Rate (Average of the above two)

```python

reduced_school_summ_df = school_summ_df[["Per Student Budget","Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]]
spending_group1_df = reduced_school_summ_df.loc[reduced_school_summ_df["Per Student Budget"] < 585,:]
spending_group1_df["group"] = "<$585"

spending_group2_df = reduced_school_summ_df.loc[(reduced_school_summ_df["Per Student Budget"] >= 585) & (reduced_school_summ_df["Per Student Budget"] < 630),:]
spending_group2_df["group"] = "$585-630"

spending_group3_df = reduced_school_summ_df.loc[(reduced_school_summ_df["Per Student Budget"] >= 630) & (reduced_school_summ_df["Per Student Budget"] < 645),:]
spending_group3_df["group"] = "$630-645"

spending_group4_df = reduced_school_summ_df.loc[reduced_school_summ_df["Per Student Budget"] >= 645,:]
spending_group4_df["group"] = ">=$645"

spending_group1_avg = round(spending_group1_df[["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]].mean(),2)
spending_group2_avg = round(spending_group2_df[["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]].mean(),2)
spending_group3_avg = round(spending_group3_df[["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]].mean(),2)
spending_group4_avg = round(spending_group4_df[["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]].mean(),2)

```

```python
score_by_spending = pd.DataFrame({" < $585 " : spending_group1_avg,
                           "$585-630" : spending_group2_avg,
                           "$630-645" : spending_group3_avg,
                           " >= $645" : spending_group4_avg
    
})

score_by_spending_df = score_by_spending.transpose()

score_by_spending_df
```

* Using binning

```python
bins = [0, 585, 630, 645, 680]

groups = [" < $585", "$585-630","$630-645"," >= $645"]

school_summ_df["Spending Ranges (Per Student)"] = pd.cut(school_summ_df["Per Student Budget"],bins,labels=groups,include_lowest=True)

score_by_spending = school_summ_df.groupby(school_summ_df["Spending Ranges (Per Student)"])

score_by_spending["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"].max()
```

# Scores by School Size

* Perform the same operations as above, based on school size.

```python
reduced_size_summ_df = school_summ_df[["Total Students","Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]]
size_group1_df = reduced_size_summ_df.loc[reduced_size_summ_df["Total Students"] < 1000,:]
size_group1_df["Size"] = " Small(<1000)"

size_group2_df = reduced_size_summ_df.loc[(reduced_size_summ_df["Total Students"] >= 1000) & (reduced_size_summ_df["Total Students"] < 2000),:]
size_group2_df["Size"] = "Medium(1000-2000)"

size_group3_df = reduced_size_summ_df.loc[(reduced_size_summ_df["Total Students"] >= 2000) & (reduced_size_summ_df["Total Students"] < 5000),:]
size_group3_df["group"] = "Large(2000-5000)"

size_group1_df = size_group1_df[["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]].mean()
size_group2_df = size_group2_df[["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]].mean()
size_group3_df = size_group3_df[["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]].mean()

```

```python
score_by_size = pd.DataFrame({ " Small(<1000)" : size_group1_df ,
                               "Medium(1000-2000)" : size_group2_df ,
                               "Large(2000-5000)" : size_group3_df    
})

score_by_size_df = score_by_size.transpose()

score_by_size_df
```

# Using binning method

```python
bins = [0, 1000, 2000, 5000]
groups = ["Small(<1000)","Medium(1000-2000)","Large(2000-5000)"]

school_summ_df["School Size"] =  pd.cut(school_summ_df["Total Students"], bins, labels=groups,include_lowest=True)

score_by_size_df = school_summ_df.groupby(school_summ_df["School Size"])
score_by_size_df["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"].max()
```
