```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

# District Summary

    Calculate the total number of schools

    Calculate the total number of students

    Calculate the total budget

    Calculate the average math score

    Calculate the average reading score

    Calculate the percentage of students with a passing math score (70 or greater)

    Calculate the percentage of students with a passing reading score (70 or greater)

    Calculate the percentage of students who passed math and reading (% Overall Passing)

    Create a dataframe to hold the above results

    Optional: give the displayed data cleaner formatting

```python
total_school = len(school_data_complete["school_name"].unique())
total_students = len(school_data_complete["student_name"])

total_budget = (school_data["budget"]).sum()

average_mathscore = round((school_data_complete["math_score"]).mean(),6)

average_readingscore = round((school_data_complete["reading_score"]).mean(),6)

passing_percent_maths = round((len(school_data_complete.loc[school_data_complete["math_score"] >= 70,:])/total_students)*100,6)

passing_percent_reading = round((len(school_data_complete.loc[school_data_complete["reading_score"] >= 70,:])/total_students)*100,6)

passing_percent_overall = round((len(school_data_complete.loc[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70),:])/total_students)*100,6)
```

```python
school_df = pd.DataFrame({ "Total Schools" : [total_school],
                          "Total Students" : [total_students],
                          "Total Budget" : [total_budget],
                          "Average Math Score" : [average_mathscore],
                          "Average Reading Score" : [average_readingscore],
                          "% Passing Math" : [passing_percent_maths],
                          "% Passing Reading" : [passing_percent_reading],
                          "% Overall Passing" : [passing_percent_overall]
    
    
})
school_df
```

# School Summary

## Create an overview table that summarizes key metrics about each school, including:

    School Name
    School Type
    Total Students
    Total School Budget
    Per Student Budget
    Average Math Score
    Average Reading Score
    % Passing Math
    % Passing Reading
    % Overall Passing (The percentage of students that passed math and reading.)
    Create a dataframe to hold the above results

```python
reduce_school_df = school_data_complete[["school_name","type","student_name","reading_score","math_score","size","budget"]]
reduce_school_df.head()
```

```python
grouped_school_df = reduce_school_df.groupby(["school_name","type"])
grouped_school_df.head()
```

```python
ttl_students = grouped_school_df["student_name"].count()

ttl_budget = grouped_school_df["budget"].mean()

ttl_math_score = grouped_school_df["math_score"].sum()
ttl_reading_score = grouped_school_df["reading_score"].sum()

math_passing_df = reduce_school_df.loc[(reduce_school_df["math_score"] >= 70), :]
grouped_math_df = math_passing_df.groupby(["school_name","type"])
ttl_math_pass_cnt = grouped_math_df["student_name"].count()

reading_passing_df = reduce_school_df.loc[(reduce_school_df["reading_score"] >= 70), :]
grouped_reading_df = reading_passing_df.groupby(["school_name","type"])
ttl_reading_pass_cnt = grouped_reading_df["student_name"].count()

overall_passing_df = reduce_school_df.loc[(reduce_school_df["math_score"] >= 70) & (reduce_school_df["reading_score"] >= 70) , :]
grouped_overall_passing_df = overall_passing_df.groupby(["school_name","type"])
ttl_overall_pass_cnt = grouped_overall_passing_df["student_name"].count()


school_summ_df = pd.DataFrame({"Total Students" : ttl_students,
                                "Total Budget" : ttl_budget,
                              "Total Maths Score":ttl_math_score,
                              "Total Reading Score":ttl_reading_score,
                               "Total Math Pass Count":ttl_math_pass_cnt,
                               "Total Reading Pass Count":ttl_reading_pass_cnt,
                               "Total Overall Pass Count":ttl_overall_pass_cnt
                              })

school_summ_df["Per Student Budget"]  = school_summ_df["Total Budget"]/school_summ_df["Total Students"]
school_summ_df["Average Math Score"] = school_summ_df["Total Maths Score"]/school_summ_df["Total Students"]
school_summ_df["Average Reading Score"] = school_summ_df["Total Reading Score"]/school_summ_df["Total Students"]
school_summ_df["% Passing Math"] = (school_summ_df["Total Math Pass Count"]/school_summ_df["Total Students"])*100
school_summ_df["% Passing Reading"] = school_summ_df["Total Reading Pass Count"]/school_summ_df["Total Students"]*100
school_summ_df["% Overall Passing"] = school_summ_df["Total Overall Pass Count"]/school_summ_df["Total Students"]*100

school_summ_df = school_summ_df[["Total Students","Total Budget","Per Student Budget","Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]]
school_summ_df
```

# Top Performing Schools (By % Overall Passing)


* Sort and display the top five performing schools by % overall passing.

```python
pass_summ_sorted = school_summ_df.sort_values(["% Overall Passing"],ascending=False).head(5)
pass_summ_sorted
```

# Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
pass_summ_sorted = school_summ_df.sort_values(["% Overall Passing"]).head(5)
pass_summ_sorted
```
